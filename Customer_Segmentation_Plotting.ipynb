```python
#Heatmap for X without target
plt.figure(figsize=(30,20))
plt.xticks(rotation=90)
plt.title("Correlation Map",fontsize=20)
sns.heatmap(foodDf.corr().round(2),vmin=-1,vmax=1,annot=True)
```

```python
plt.figure(1, figsize=(12,4))
n = 0
for x in ['Age', 'Income', 'Spending']:
  n+=1
  plt.subplot(1,3,n)
  plt.subplots_adjust(hspace= 0.5, wspace=0.5)
  sns.distplot(foodDf[x], bins = 20)
  plt.title(' MarketingCampaign Distplot of {}'.format(x))
plt.show()
```

```python
plt.figure(1, figsize=(15,7))
n = 0
for cols in ['Catalog', 'Income', 'Family_Size']:
  n+=1
  plt.subplot(1,3,n)
  sns.set(style = 'whitegrid')
  plt.subplots_adjust(hspace= 0.5, wspace=0.5)
  sns.violinplot(x = cols,y = 'Age', data = foodDf)
  plt.ylabel('Gender' if n==1 else '')
  plt.title('Violin Plot')
plt.show()
```

```python
# What is the distribution of the age?

age_18_25 = foodDf.Age[(foodDf.Age >= 18) & (foodDf.Age <= 25)]
age_26_35 = foodDf.Age[(foodDf.Age >= 26) & (foodDf.Age <= 35)]
age_36_45 = foodDf.Age[(foodDf.Age >= 36) & (foodDf.Age <= 45)]
age_46_55 = foodDf.Age[(foodDf.Age >= 46) & (foodDf.Age <= 55)]
age_above_55 = foodDf.Age[(foodDf.Age >= 56)]

agex = ['18-25', '26-35', '36-45','46-55','55+']
agey = [len(age_18_25.values),len(age_26_35.values),len(age_36_45.values),len(age_46_55.values),len(age_above_55.values)]

plt.figure(figsize = (15,6))
sns.barplot(x = agex, y = agey , palette='mako')
plt.title('')
plt.xlabel('Age')
plt.ylabel('Number of Customer')
plt.show()
```
