# Feature Engineering

## Creating a feature how long the customers are part of the company

```python
foodDf["Dt_Customer"] = pd.to_datetime(foodDf["Dt_Customer"])
dates = []
for i in foodDf["Dt_Customer"]:
    i = i.date()
    dates.append(i)  
#Dates of the newest and oldest recorded customer
print("The newest customer's enrollment date in the records:",max(dates))
print("The oldest customer's enrollment date in the records:",min(dates))
```

```python
foodDf["Dt_Customer"]
```

```python
youngest= foodDf["Dt_Customer"].min()
```

```python
oldest = foodDf["Dt_Customer"].max()
```

```python
#Created a feature "Since_When"
days = []
day1 = max(dates) #taking it to be the newest customer
for i in dates:
    x = day1 - i
    days.append(x)
foodDf["Since_When"] = days
```

```python
foodDf["Since_When"]
```

```python
# Getting only the digits since how many days the people are customers
foodDf["Since_When"] = foodDf["Since_When"].astype(str).str.extract(r'(^\d+)')
```

```python
foodDf
```

## Renaming and replacing values

```python
# Creating just Partner or Alone as unique values
foodDf["Marital_Status"]=foodDf["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})

#Segmenting education levels in three groups - Undergraduate - Graduate - Postgraduate
foodDf["Education"]=foodDf["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})
```

## Renaming columns

```python
#Renaming products
foodDf=foodDf.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})

#Renaming places
foodDf = foodDf.rename(columns={'NumWebPurchases': "Web",'NumCatalogPurchases':'Catalog','NumStorePurchases':'Store'})
```

## Creating new features out of the old ones

```python
#Age of customer today 
foodDf["Age"] = 2021-foodDf["Year_Birth"]

#Total spendings on various items
foodDf["Spending"] = foodDf["MntWines"]+ foodDf["MntFruits"]+ foodDf["MntMeatProducts"]+ foodDf["MntFishProducts"]+ foodDf["MntSweetProducts"]+ foodDf["MntGoldProds"]

#Feature indicating total children living in the household
foodDf["Children"]=foodDf["Kidhome"]+foodDf["Teenhome"]

#Creating family_size to tell the number of members in every household
foodDf["Family_Size"] = foodDf["Marital_Status"].replace({"Alone": 1, "Partner":2})+ foodDf["Children"]

#Feature pertaining parenthood
foodDf["Parent"] = np.where(foodDf.Children> 0, 1, 0)
```

## Dropping redundant features

```python
#Dropping some of the redundant features
to_drop = ["Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID"]
foodDf = foodDf.drop(to_drop, axis=1)
```
